```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import csv
```

```python
from sklearn.linear_model import SGDClassifier,LogisticRegression
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.metrics import log_loss
from sklearn.ensemble import AdaBoostClassifier,RandomForestClassifier, GradientBoostingClassifier
```

```python
df=pd.read_csv('train.csv',parse_dates=['Dates'])#878049
```

```python
X_X=df.X
X_Y=df.Y
```

```python
lc1=preprocessing.LabelEncoder()
X_DayofWeek = pd.DataFrame(lc1.fit_transform(df.DayOfWeek))
lc2=preprocessing.LabelEncoder()
X_PdDistrict = pd.DataFrame(lc2.fit_transform(df.PdDistrict))
```

```python
year = df.Dates.dt.year
month = df.Dates.dt.month
day = df.Dates.dt.day
hour = df.Dates.dt.hour
```

```python
labelcoder=preprocessing.LabelEncoder()
```

```python
X_Category=labelcoder.fit_transform(df.Category)
Y_Train=pd.DataFrame(np.array(X_Category))
```

```python
X_Train=pd.concat([year,month,day,hour,X_Time,X_DayofWeek,X_PdDistrict,X_X,X_Y,X_StreetCorner,X_Block],axis=1)
```
